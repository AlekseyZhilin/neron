```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
from sklearn.metrics import r2_score, mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

tf.random.set_seed(1)
%matplotlib inline
```

```python
# Загрузка данных
dataset = load_boston()
```

```python
X = dataset.data
Y = dataset.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=1)
```

```python
X_train.shape
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Flatten, Dense
from tensorflow.keras.optimizers import Adam, SGD
from tensorflow.keras.models import Model
from tensorflow.keras.losses import mean_squared_error
```

```python
%%time
tf.random.set_seed(1)
# будем сохранять модели в словарь
plt.figure(figsize=(16, 7))
models = {}
models_res = np.array([0,0])

num_epochs = 8
batch_size = 50
laers = [1, 2, 3, 4, 5, 10, 50, 100]
# вектор для вывода результатов
epoch = np.arange(num_epochs+1)

for level in laers:
    print('=' * 50 + f'Модель laers_{level}')
    # создаём модель
    inputs = Input(shape=(X_train.shape[1]), name= 'features') 
    x = Flatten()(inputs)
    for i in range(level):
        x = Dense(64, activation='relu', name=f'dense_{i}')(x)


    outputs = Dense(1, activation='relu', name='predictions')(x) 
    model_i = Model(inputs=inputs, outputs=outputs, name='laers_'+str(level))
    
    # компилируем model_i
    model_i.compile(optimizer=Adam(), loss=mean_squared_error, metrics=['mae'])
    
    # вычисляем ошибку для modeli без обучения
    h0_train = model_i.evaluate(X_train, y_train, verbose=0)
    h0_val = model_i.evaluate(X_test, y_test, verbose=0)
    
    # обучаем модель
    h = model_i.fit(X_train, y_train,
          epochs=num_epochs,
          batch_size=batch_size,
          validation_data=(X_test, y_test))
    
    # записываем результаты в список
    test_loss, test_mae = model_i.evaluate(X_test, y_test, verbose=0)
    models_res = np.vstack((models_res, np.array([str(model_i.name), test_mae])))
    
    # записываем обученную модель в словарь
    models[str(model_i.name)] = model_i
    
    # картинки
    plt.plot(epoch, np.concatenate((np.array([0]), np.array(h.history['mae'])), axis=0),
             '-',
             label=str(model_i.name) + ' train')
    plt.plot(epoch, np.concatenate((np.array([0]), np.array(h.history['val_mae'])), axis=0),
             '--',
             label=str(model_i.name) + ' val')
    
    
plt.legend()
plt.title('mae при различных количествах слоёв в нейронной сети')
plt.xlabel('epoch')
plt.ylabel('mae')
#plt.ylim((15, 25))
plt.show()
```

### При увеличении числа слоёв до 10, метрика mae улучшается на train и test. Однако, дальнейшее увеличением слоёв, приводит к ухудшению метрики качества. Возрастание числа эпох, так же вносит существенное улучшение метрики mae. Для большинства случаев, увеличение эпох, даёт лучший результат оптимизации mae, по сравнению с увеличением слоёв.

### Изменение количества нейронов в слоях (постепенное уменьшение, с учётом ранее полученого оптимального количества в первом слое), так же привело к существенному улучшеню метрики качества. Однако при сравнении с предыдущими результатами при равном количестве слоёв и эпох (10 и 8 соответственно), модель с одинаковым количеством нейронов в слоях (64) показала немного лучший результат

```python
%%time
tf.random.set_seed(1)
# будем сохранять модели в словарь
plt.figure(figsize=(16, 7))
models_l_n = {}
models_res_l_n = np.array([0,0])

num_epochs = 8
batch_size = 50
nerons = [128, 64]
# вектор для вывода результатов
epoch = np.arange(num_epochs+1)

for ner in nerons:
    print('=' * 50 + f'Модель ner_{ner}')
    # создаём модель
    inputs = Input(shape=(X_train.shape[1]), name= 'features') 
    x = Flatten()(inputs)
    for i in range(10):
        if i % 2 == 0:
            x = Dense(ner / (i + 1), activation='relu', name=f'dense_{i}')(x)
        else:
            x = Dense(ner / (i + 1), activation='linear', name=f'dense_{i}')(x)


    outputs = Dense(1, activation='relu', name='predictions')(x) 
    model_i = Model(inputs=inputs, outputs=outputs, name='ner_'+str(ner))
    
    # компилируем model_i
    model_i.compile(optimizer=Adam(), loss=mean_squared_error, metrics=['mae'])
    
    # вычисляем ошибку для modeli без обучения
    h0_train = model_i.evaluate(X_train, y_train, verbose=0)
    h0_val = model_i.evaluate(X_test, y_test, verbose=0)
    
    # обучаем модель
    h = model_i.fit(X_train, y_train,
          epochs=num_epochs,
          batch_size=batch_size,
          validation_data=(X_test, y_test))
    
    # записываем результаты в список
    test_loss, test_mae = model_i.evaluate(X_test, y_test, verbose=0)
    models_res_l_n = np.vstack((models_res_l_n, np.array([str(model_i.name), test_mae])))
    
    # записываем обученную модель в словарь
    models_l_n[str(model_i.name)] = model_i
    
    # картинки
    plt.plot(epoch, np.concatenate((np.array([0]), np.array(h.history['mae'])), axis=0),
             '-',
             label=str(model_i.name) + ' train')
    plt.plot(epoch, np.concatenate((np.array([0]), np.array(h.history['val_mae'])), axis=0),
             '--',
             label=str(model_i.name) + ' val')
    
    
plt.legend()
plt.title('mae при различных количествах нейронов в слое нейронной сети')
plt.xlabel('epoch')
plt.ylabel('mae')
#plt.ylim((15, 25))
plt.show()
```
