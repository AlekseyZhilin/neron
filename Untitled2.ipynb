```python
import keras
import matplotlib.pyplot as plt
from keras.datasets import fashion_mnist
import numpy as np
import tensorflow as tf
tf.random.set_seed(1)
```

```python
# Загрузка данных
(X_train, y_train_labels), (X_test, y_test_labels) = fashion_mnist.load_data()

print(X_train.shape, X_test.shape)
print(y_train_labels.shape)
```

```python
count_val = np.max(y_train_labels) + 1 if np.min(y_train_labels) == 0 else np.max(y_train_labels)
```

```python
print(f'Количество разных элементов в датафрейме = {count_val}')
```

```python
# репрезентация данных
plt.figure(figsize=(16, 6))
for i in range(30):
    plt.subplot(3, 10, i + 1)
    plt.imshow(X_train[i,:].reshape([28, 28]))
```

```python
# Нормировка
X_min, X_max = X_train.min(), X_train.max()
print(f'min val = {X_min}, max val = {X_max}')
```

```python
X_train = X_train / X_max
X_test = X_test / X_max
print(f'min val = {X_train.min()}, max val = {X_train.max()}')
```

```python
# Вытягивание признаков
from keras.utils.np_utils import to_categorical

y_train = to_categorical(y_train_labels)
y_test = to_categorical(y_test_labels)

y_train[:5]
```

```python
X_train = X_train.reshape((-1, 28 * 28))
X_test = X_test.reshape((-1, 28 * 28))

X_train.shape, X_test.shape
```

```python
from keras.layers import Dense
from keras.models import Sequential
```

```python
%%time


# будем сохранять модели в словарь
models = {}


plt.figure(figsize=(16, 7))
colors = ['g', 'b', 'r', 'y']

num_epochs = 6
batch_size = 100

# вектор для вывода результатов
epoch = np.arange(num_epochs+1)

# будем изменять оптимизаторы
i=-1
for ner in [[16, 8], [32, 16], [64, 32], [128, 64]]:
    i+=1
    print('=' * 50 + str(ner[0]) + ', ' + str(ner[1]) + '=' * 50)
    # создаем рабочую модель  model_i, куда будем загружать эти веса (она должна быть идентичной структуры)
    model_i = Sequential([
        Dense(ner[0], activation='relu', input_shape=(784,)),
        Dense(ner[1], activation='relu'),
        Dense(10, activation='sigmoid'),
    ])
    
    # загружаем веса model_i.load_weights('model.h5')
    #model_i.load_weights('model_weights.h5')
    
    # компилируем model_i с одним из оптимизаторов
    model_i.compile(
        optimizer='Adam',
        loss='categorical_crossentropy',
        metrics=['accuracy'],
        )
    
    # вычисляем ошибку для modeli без обучения
    h0_train = model_i.evaluate(X_train, y_train, verbose=0)
    h0_val = model_i.evaluate(X_test, y_test, verbose=0)

    # проводим обучение модели 
    h = model_i.fit(X_train, y_train,
                    epochs=num_epochs,
                    batch_size=batch_size,
                    validation_data=(X_test, y_test),
                    verbose=1)
    
    # записываем обученную модель в словарь 
    models[str(ner[0]) + ', ' + str(ner[1])] = model_i

    # картинки
    plt.plot(epoch, np.concatenate((np.array([0]), np.array(h.history['accuracy'])), axis=0),
             '-', c=colors[i],
             label=str(ner[0]) + ', ' + str(ner[1]) + ' train')
    plt.plot(epoch, np.concatenate((np.array([0]), np.array(h.history['val_accuracy'])), axis=0),
             '--', c=colors[i],
             label=str(ner[0]) + ', ' + str(ner[1]) + ' val')
    
plt.legend()
plt.title('accuracy при различных количествах нейронов в двухслойной сети')
plt.xlabel('epoch')
plt.ylabel('accuracy')
#plt.yscale(value="log")
plt.ylim((0.77, 0.9))
plt.show()
```

```python
# Предсказание модели
N = 5
model_i = models['128, 64']
y_pred = model_i.predict(X_test[:N, :])
y_pred_class = np.argmax(y_pred, axis=1)
plt.figure(figsize=(3 * N, 5))

for i in range(N):
  plt.subplot(1, N, i+1)
  plt.imshow(X_test[i,:].reshape(28, 28))
  plt.xlabel(str(y_pred_class[i]))
plt.show()  
```

## Ответ на 1.2 вопрос

### Видно, что при увеличение количества слоёв, метрика accuracy снижается и на train и на test. Целесообразно использовать 2 или 3 слоя (наилучшая метрика 0.8702 и 0.8720 на тесте)

```python
%%time

# будем сохранять модели в словарь
models_opt = {}

plt.figure(figsize=(16, 7))
colors = ['g', 'b', 'r', 'y']

num_epochs = 6
batch_size = 100

# вектор для вывода результатов
epoch = np.arange(num_epochs+1)

# будем изменять оптимизаторы
for i, i_optim in enumerate([keras.optimizers.gradient_descent_v2.SGD(),
                             keras.optimizers.adagrad_v2.Adagrad(),
                             keras.optimizers.rmsprop_v2.RMSprop(),
                             keras.optimizers.adam_v2.Adam()]):
    
    
    model_i = Sequential([
        Dense(64, activation='relu', input_shape=(784,)),
        Dense(32, activation='relu'),
        Dense(10, activation='sigmoid'),
    ])
    
    # компилируем model_i с одним из оптимизаторов
    model_i.compile(
        optimizer=i_optim,
        loss='categorical_crossentropy',
        metrics=['accuracy'],
        )
    
    # вычисляем ошибку для modeli без обучения
    h0_train = model_i.evaluate(X_train, y_train, verbose=0)
    h0_val = model_i.evaluate(X_test, y_test, verbose=0)

    # проводим обучение модели 
    h = model_i.fit(X_train, y_train,
                    epochs=num_epochs,
                    batch_size=batch_size,
                    validation_data=(X_test, y_test),
                    verbose=1)

    # записываем обученную модель в словарь
    models_opt[i_optim.get_config()['name']] = model_i

    # картинки
    plt.plot(epoch, np.log([h0_train[0]] + h.history['loss']),
             '-', c=colors[i],
             label=model_i.optimizer.get_config()['name'] + ' train')
    plt.plot(epoch, np.log([h0_val[0]] + h.history['val_loss']),
             '--', c=colors[i],
             label=model_i.optimizer.get_config()['name'] + ' val')
    print('=' * 20)

plt.legend()
plt.title('нейросети от оптимизатора log(loss)')
plt.xlabel('epoch')
plt.ylabel('log(loss)')
```
