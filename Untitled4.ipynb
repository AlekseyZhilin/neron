```python
import tensorflow.keras as keras
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras import models, layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

import os, ssl
import tensorflow as tf
tf.random.set_seed(1)
%matplotlib inline
```

```python
classes = ['самолет', 'автомобиль', 'птица', 'кот', 'олень', 'собака', 'лягушка', 'лошадь', 'корабль', 'грузовик']
```

```python
# Загрузка данных
# разделение тренировочной и тестовой выборки
#if (not os.environ.get('PYTHONHTTPSVERIFY', '') and getattr(ssl, '_create_unverified_context', None)):
#    ssl._create_default_https_context = ssl._create_unverified_context
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
```

```python
# установка параметров нейросети
batch_size = 500
num_classes = 10
epochs = 5
# преобразование матрицы чисел в бинарную матрицу
y_train = to_categorical(y_train, num_classes)
y_test = to_categorical(y_test, num_classes)
```

```python
# Нормализация данных
print(f'X_train.max = {X_train.max()}, X_test.max = {X_test.max()}')

X_train = X_train / X_train.max()
X_test = X_test / X_test.max()
```

```python
# Просмотр загруженного изображения
N = 1

plt.imshow(X_test[N])
plt.title(classes[np.argmax(y_test[N,:])])
plt.show()
```

```python
# конфигурирование слоев нейросети
model = Sequential()

# слои нейросети отвественные за свертку и max-pooling
model.add(Conv2D(32, (3, 3), padding='same', input_shape=X_train.shape[1:]))
model.add(Activation('relu'))
model.add(Conv2D(32, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Conv2D(64, (3, 3), padding='same'))
model.add(Activation('relu'))
model.add(Conv2D(64, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

# полносвязные слои нейронной сети
model.add(Flatten())
model.add(Dense(500))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(num_classes))
model.add(Activation('softmax'))

model.summary()
```

```python
# компиляция модели
model.compile(loss='categorical_crossentropy',
              optimizer='SGD',
              metrics=['accuracy'])
```

```python
# Обучение модели
hist = model.fit(X_train, y_train,
                 epochs=epochs,
                 batch_size=batch_size,
                 validation_data=(X_test, y_test),
                 verbose=1)
```

```python
# проверка работы обученной модели
test_loss, test_metric = model.evaluate(X_test, y_test)
print(f"Test loss {test_loss:.4f}, accuracy {test_metric:.2f}%")
```
