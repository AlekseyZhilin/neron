```python
import tensorflow.keras as keras
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras import models, layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D, AveragePooling2D
import numpy as np
import matplotlib.pyplot as plt

#import os, ssl
import tensorflow as tf
tf.random.set_seed(1)
%matplotlib inline
```

```python
classes = ['самолет', 'автомобиль', 'птица', 'кот', 'олень', 'собака', 'лягушка', 'лошадь', 'корабль', 'грузовик']
```

```python
# Загрузка данных
# разделение тренировочной и тестовой выборки
#if (not os.environ.get('PYTHONHTTPSVERIFY', '') and getattr(ssl, '_create_unverified_context', None)):
#    ssl._create_default_https_context = ssl._create_unverified_context
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# установка параметров нейросети
batch_size = 512
num_classes = 10
# Количество эпох для обучения снижено из-за большого времени обучения 
epochs = 3
# преобразование матрицы чисел в бинарную матрицу
y_train = to_categorical(y_train, num_classes)
y_test = to_categorical(y_test, num_classes)
```

```python
# Нормализация данных
print(f'X_train.max = {X_train.max()}, X_test.max = {X_test.max()}')

X_train = X_train / X_train.max()
X_test = X_test / X_test.max()
```

```python
# Просмотр загруженного изображения
N = 1

plt.imshow(X_test[N])
plt.title(classes[np.argmax(y_test[N,:])])
plt.show()
```

```python
%%time
tf.random.set_seed(1)
# будем сохранять модели в словарь
models = {}
models_res = np.array([0,0])

# конфигурирование слоев нейросети
for conv in [8, 6, 5, 4]:
    print('=' * 50 + f'Модель conv_{conv}')
    model = Sequential(name='conv_' + str(conv))
    # слои нейросети отвественные за свертку и max-pooling
    model.add(Conv2D(32, (conv, conv), padding='same', activation='relu',input_shape=X_train.shape[1:]))
    model.add(AveragePooling2D(pool_size=(2, 2)))
    model.add(Conv2D(32, (conv - 1, conv - 1), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Dropout(0.25))
    model.add(Conv2D(64, (3, 3), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    # полносвязные слои нейронной сети
    model.add(Flatten())
    model.add(Dense(512, activation='relu'))
    model.add(Dense(num_classes))
    model.add(Activation('softmax'))
    
    # компиляция модели
    model.compile(loss='categorical_crossentropy',
                  optimizer='Adam',
                  metrics=['accuracy'])
    
    # Обучение модели
    hist = model.fit(X_train, y_train,
                 epochs=epochs,
                 batch_size=batch_size,
                 validation_data=(X_test, y_test),
                 verbose=1)
    
    # проверка работы обученной модели
    test_loss, test_metric = model.evaluate(X_test, y_test)
    print(f"Test loss {test_loss:.4f}, accuracy {test_metric:.2f}%")
    
    # записываем обученную модель в словарь
    models[str(model.name)] = model
    models_res = np.vstack((models_res, np.array([str(model.name), test_metric])))
```

```python
models_res = models_res[1:]
models_res
```

### Однозначного вывода, что увеличение размера свёртки влияет на увеличение метрики качества accuracy, сделать нельзя. Обратное утверждение, то же неверно.
### Видны разнонаправленные тендениции. Необходимо экспериментально подбирать размер.
### В первом слое оптимально выбрать 5*5, с последующим уменьшением

```python
%%time
tf.random.set_seed(1)
# будем сохранять модели в словарь
models_ = {}
models_res_ = np.array([0,0])

# конфигурирование слоев нейросети
for feat in [8, 16, 32, 64]:
    print('=' * 50 + f'Модель feat_{feat}')
    model = Sequential(name='feat_' + str(feat))
    # слои нейросети отвественные за свертку и max-pooling
    model.add(Conv2D(feat, (5, 5), padding='same', activation='relu',input_shape=X_train.shape[1:]))
    model.add(AveragePooling2D(pool_size=(2, 2)))
    model.add(Conv2D(feat*2, (4, 4), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Dropout(0.25))
    model.add(Conv2D(feat*4, (3, 3), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    # полносвязные слои нейронной сети
    model.add(Flatten())
    model.add(Dense(512, activation='relu'))
    model.add(Dense(num_classes))
    model.add(Activation('softmax'))
    
    # компиляция модели
    model.compile(loss='categorical_crossentropy',
                  optimizer='Adam',
                  metrics=['accuracy'])
    
    # Обучение модели
    hist = model.fit(X_train, y_train,
                 epochs=epochs,
                 batch_size=batch_size,
                 validation_data=(X_test, y_test),
                 verbose=1)
    
    # проверка работы обученной модели
    test_loss, test_metric = model.evaluate(X_test, y_test)
    print(f"Test loss {test_loss:.4f}, accuracy {test_metric:.2f}%")
    
    # записываем обученную модель в словарь
    models_[str(model.name)] = model
    models_res_ = np.vstack((models_res_, np.array([str(model.name), test_metric])))
```

```python
models_res_ = models_res_[1:]
models_res_
```

### Последовательное увеличение слоёв повышает метрику качества accuracy. 
### Причём, с увеличением количества слоёв, метрика увеличивается, но так же, существенно увеличивается и время обучения
